# Projekt 1 - Filmratings

```python
import pandas as pd
```

# 1. Läs in all användardata
Läser in först all vår data där vi sparat den

```python
ratingData = pd.read_csv("u.data", sep="\t", names=["userId", "itemId", "rating", "timestamp"])
#ratingData

movieData = pd.read_csv("u.item", sep="|", encoding="latin-1", names=[
                                                                    "movieId",
                                                                    "movieTitle",
                                                                    "releaseDate",
                                                                    "videoReleaseDate",
                                                                    "IMDbURL",
                                                                    "unknown",
                                                                    "Action",
                                                                    "Adventure",
                                                                    "Animation",
                                                                    "Children\'s",
                                                                    "Comedy",
                                                                    "Crime",
                                                                    "Documentary",
                                                                    "Drama",
                                                                    "Fantasy",
                                                                    "Film-Noir",
                                                                    "Horror",
                                                                    "Musical",
                                                                    "Mystery",
                                                                    "Romance",
                                                                    "Sci-Fi",
                                                                    "Thriller",
                                                                    "War",
                                                                    "Western"])
#movieData

userData = pd.read_csv("u.user", sep="|", names=["userId","age","gender","occupation","zipCode"])
```

```python
#userData.head()
#movieData.head()
#ratingData.head()

#userData.dtypes
#movieData.dtypes
#ratingData.dtypes

#userData.describe
#movieData.describe
#ratingData.describe
```

# 2. Välj data

#### - Visa endast kolumnerna kön, ålder och yrke

```python
# Skapa en ny DataFrame, sortUserData, där vi bara tar kolumnerna 
# "age", "gender" och "occupation" från den ursprungliga DataFramen
sortUserData = userData[["age", "gender", "occupation"]]
sortUserData
```

```python
ratingData
```

#### - Gruppera ratings enligt film

```python
# Skapar en ny DataFrame, sortRatingData, där vi sätter filmens id som index

sortRatingData = ratingData.set_index('itemId')

# Vi sorterar den, alla filmers gradering i ordning
sortRatingData.sort_index(inplace=True)

sortRatingData
```

Bara för skojs skull här har vi satt ihop en DataFrame där den totala betygsättning presenteras, inget annat speciellt, skippa denna del

```python
movieTotalRatings = sortRatingData[["rating"]]
movieTotalRatings = movieTotalRatings.groupby('itemId').sum()
movieTotalRatings = movieTotalRatings.sort_values('rating', ascending=False)
movieTotalRatings
```

```python
movieTotalRatings.sort_values('itemId')
```

# 3. Filtrera data

```python
'''
Ändrar inte ursprungliga DataFramen.
Här ser vi också att det är främst män med yrket tekniker, faktume, bara EN kvinna i datasetet har yrket tekniker
'''
userData[userData.occupation.str.contains('technician')]
```

```python
# vi ändrar/skapar inte heller någon ny DataFrame 
# för det vi vill åstadkomma här, vi vill bara presenter
# den data som vi vill

userData[ (userData.gender == 'M') & (userData.age >= 40)].sort_values('age')
```

# 4. Utforska data
### Medelålder på alla författare

```python
userData[userData.occupation == "writer"].age.mean()
```

### Medelrating per film

```python
sortRatingData.groupby('itemId').rating.mean()
```

### Top 10 filmer med >=40 ratings

Fast det stod i beskrivningen över 40 ratings tycker jag vi kan inkludera dom med 40

```python
# skapar en ny df, vi checkar antalet av givna ratings och
# rensar resultatet genom att ta bort de som inte uppfyller kravet

fourtyRatings = sortRatingData[sortRatingData.groupby('itemId').count() >= 40].dropna()
fourtyRatings.drop(['userId', 'timestamp'], axis=1, inplace=True)
#fourtyRatings[fourtyRatings.index == 1227].count()
#fourtyRatings
```

```python
fourtyRatings = fourtyRatings.groupby('itemId').mean()
fourtyRatings.sort_values('rating', ascending=False).head(10)
```

# 5. Kombinera DataFrames
### Vanliga yrken etc.

Vi skapar en ny DataFrame där vi indexerar med de givna värden och räknar den totala mängden kön per profession

```python
# kunde göra en finare DataFrame genom att gruppera och indexera mera annorlunda
# men detta gör jobbet

pPopularity = userData.drop(["userId", "age", "zipCode"], axis=1)
pPopularity = pPopularity.set_index('occupation', 'gender') 
pPopularity["total"] = pPopularity.groupby(['occupation']).agg({'gender':['count']})
pPopularity = pPopularity.groupby(['occupation', 'gender']).agg('count')
```

```python
pPopularity.sort_values(['gender','total'], ascending=False)
```

### Kombinera u.data med u.item

```python
combinedData = pd.merge(sortRatingData, movieData, left_on='itemId', right_on='movieId')
```

```python
combinedData
```

### Vi kombinerar all data

```python
completeData = pd.merge(userData, combinedData)
```

```python
completeData
```

#### Låt oss ta en titt på de top 5 snällaste och tuffaste betygssättarna

```python
# skapa ny df med givna kolumner och gruppering

userMeanRatings = pd.DataFrame(completeData[['userId', 'age', 'gender', 'occupation', 'rating']])
userMeanRatings = userMeanRatings.groupby(['userId', 'occupation', 'gender']).mean()
userMeanRatings.sort_values('rating', ascending=False)
```

Medelåldern på både tuffaste och snällaste är relativt låga, men vi ser att medelåldern 

för de snällaste är 6 år mindre (medianen: 18). 

```python
userMeanRatings.sort_values('rating', ascending=False).head().mean()
```

```python
userMeanRatings.sort_values('rating', ascending=False).tail().mean()
```

Medelåldern på alla är t.ex. 34, så de är båda relativt, eller ganska så unga

```python
userMeanRatings.mean()
```
