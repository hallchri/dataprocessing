### Vi skapar våra toner och kombinerar dom till ett ackord

<b>Detta kundes säkert göra på bättre och snyggare sätt, men mina kunskap inom musikteori är i stort sätt icke-existerande, men jag är nöjd med mitt resultat oavsett och vänta bara tills du lyssnar på mästervärket, nästa Vivaldi!</b>

Och för att vara ärlig, jag snappade inte upp direkt något konkret angående Nyquists "juttun" och det att arrayn måste vara "minst dubbelt" sampelfrekvensen? Jag har åtminstone utgått från en sampelfrekvens på 8kHz i koden men samplar själva ackorden på 16kHz

Så alltså, att läsa källorna om musikteori (wikipedia) är nog värre än hebreiska för mig, men nog förstod jag hur vi kommer till det önskade resultatet och har naturligtvis lärt mig något angående musikteori, de e roligt! Och ursäkta på samma gång om min lösning går emot goda standarder, men låt mig vara glad för att ha uppnått resultatet! ;D 



```python
# Tidsintervall och sampelrate
t = 0.000125
s = 8000
timescale = np.zeros(s, dtype="float32") # vi skapar vår tidsskala-array med storleken 8000

# Formel för att hitta frekvensen för respektive semiton från A
# freq = A/2**(n/12)

# Här kunde man istället skapa funktioner för att på ett mera korrekt sätt konstruera diverse noter
# ...Nåväl!

A = 440           # A-notens frekvens, tycks vara standard att utgå från den
fA = A*10         # Frekvens med vår amplitud
C = fA/2**(9/12)  # Respektive noter
E = fA/2**(5/12)
G = fA/2**(2/12)

# Vi skapar tre arrays med s-antal linjära värden från 0 -> respektive notfrekvens
sinOne = np.linspace(0, C, s)
sinTwo = np.linspace(0, E, s)
sinThree = np.linspace(0, G, s)
yAll = sinOne + sinTwo + sinThree

# Därefter räknar vi ut sinuset för noterna
y1 = np.sin(sinOne)
y2 = np.sin(sinTwo)
y3 = np.sin(sinThree)

# 
combined = (y1 + y2 + y3) / 3 # Dividera med x-antal not-arrays vi kombinerar, annars följer vi inte sinusvågens natur
```

```python
# Populera arrayn med tidpunkterna för våra vågvärden för ackordet (om man kan säga så?) Terminologi är inte min starka sida :'(
for n in range(s):
    timescale[n] = 0
    timescale[n] = t
    t += 0.000125
    
timescale[:5] # Ser bra ut... KÖR BARA INTE DENNA CELL INDIVIDUELLT x-antal gånger ;D
```

```python
# Vi illustrerar våra sinusvågar i en graf

fig = plt.figure()
a = fig.add_subplot(1,2,1)
a.set_title("Individual")

waveIndex = 50

plt.plot(timescale[:waveIndex], y1[:waveIndex], "b", label="Sin1")
plt.plot(timescale[:waveIndex], y2[:waveIndex], "g", label="Sin2")
plt.plot(timescale[:waveIndex], y3[:waveIndex], "y", label="Sin3")
plt.ylim(-1.5, 1.5)
plt.xlim(0)

# Också vårt ackord vill vi se

chordIndex = 50
a = fig.add_subplot(1,2,2)
a.set_title("Combined")
plt.plot(timescale[:chordIndex], combined[:chordIndex], label="C Chord")
plt.ylim(-1.5, 1.5)
plt.xlim(0)

# Det ser bra ut!
```

```python
# Ackorden, combined, som vi skickar in dividerar vi med 5 för att få en mer "komprimerad" våg
# Samt skapar vi dom två första raderna i .csv:n som krävs för en lämplig .dat-fil
np.savetxt('cacc.csv', np.transpose([timescale, combined/5]), delimiter=' ', 
           comments="", header="; Sample Rate 16000\n; Channels 1")
```
